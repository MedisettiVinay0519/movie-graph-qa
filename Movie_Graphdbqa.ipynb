```python
# Create Movie nodes
for index, row in df.iterrows():
    graph.query(
        """
        MERGE (m:Movie {title: $title, year: $year, imdb_rating: $imdb_rating, duration: $duration, revenue_million: $revenue_million})
        """,
        params={
            "title": row['Title'],
            "year": row['Year'],
            "imdb_rating": row['IMDB_Rating'],
            "duration": row['Duration'],
            "revenue_million": row['Revenue_Million']
        }
    )

# Create Genre nodes and relationships
for index, row in df.iterrows():
    genres = row['Genre'].split(',') # Assuming genres are comma-separated
    for genre in genres:
        graph.query(
            """
            MERGE (g:Genre {name: $genre_name})
            WITH g
            MATCH (m:Movie {title: $movie_title})
            MERGE (m)-[:HAS_GENRE]->(g)
            """,
            params={"genre_name": genre.strip(), "movie_title": row['Title']}
        )

# Create Director nodes and relationships
for index, row in df.iterrows():
    graph.query(
        """
        MERGE (d:Director {name: $director_name})
        WITH d
        MATCH (m:Movie {title: $movie_title})
        MERGE (d)-[:DIRECTED]->(m)
        """,
        params={"director_name": row['Director'], "movie_title": row['Title']}
    )

# Create Actor nodes and relationships
for index, row in df.iterrows():
    actors = [row['Actor1'], row['Actor2'], row['Actor3']]
    for actor in actors:
        graph.query(
            """
            MERGE (a:Actor {name: $actor_name})
            WITH a
            MATCH (m:Movie {title: $movie_title})
            MERGE (a)-[:ACTS_IN]->(m)
            """,
            params={"actor_name": actor, "movie_title": row['Title']}
        )

# Create Language nodes and relationships
for index, row in df.iterrows():
    graph.query(
        """
        MERGE (l:Language {name: $language_name})
        WITH l
        MATCH (m:Movie {title: $movie_title})
        MERGE (m)-[:HAS_LANGUAGE]->(l)
        """,
        params={"language_name": row['Language'], "movie_title": row['Title']}
    )

# Create Country nodes and relationships
for index, row in df.iterrows():
    graph.query(
        """
        MERGE (c:Country {name: $country_name})
        WITH c
        MATCH (m:Movie {title: $movie_title})
        MERGE (m)-[:PRODUCED_IN]->(c)
        """,
        params={"country_name": row['Country'], "movie_title": row['Title']}
    )
```

```python
# ================================
# 1️⃣ Create Indexes for faster queries
graph.query("""CREATE INDEX movie_title_index IF NOT EXISTS FOR (m:Movie) ON (m.title)""")
graph.query("""CREATE INDEX actor_name_index IF NOT EXISTS FOR (a:Actor) ON (a.name)""")
graph.query("""CREATE INDEX director_name_index IF NOT EXISTS FOR (d:Director) ON (d.name)""")
graph.query("""CREATE INDEX genre_name_index IF NOT EXISTS FOR (g:Genre) ON (g.name)""")
graph.query("""CREATE INDEX country_name_index IF NOT EXISTS FOR (c:Country) ON (c.name)""")
graph.query("""CREATE INDEX language_name_index IF NOT EXISTS FOR (l:Language) ON (l.name)""")

# ================================
# The LOAD CSV part is removed as it requires specific database configuration
# You can use the code in cell sTa17YUNCDJN to load data from the pandas DataFrame
# or use other methods like neo4j-admin import or a client application.
```
